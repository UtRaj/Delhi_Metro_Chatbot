```python
import numpy as np
import pandas as pd
import decimal

# Importing regex and random libraries
import re
import random

class AlienBot:
    # Potential negative responses
    negative_responses = ("no", "nope", "nah", "naw", "not a chance", "sorry")
    # Keywords for exiting the conversation
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")
    # Random starter questions
    random_questions = (
        "How can I help you?\nWrite book for booking tickets!!!\nWrite time to know the Metro time\nWrite help for any help!!!\nWrite exit to quit chatbot\n",
        "How may I assist you?\nWrite book for booking tickets!!!\nWrite time to know the Metro time\nWrite help for any help!!!\nWrite exit to quit chatbot\n",
        "What assistance can I provide you?\nWrite book for booking tickets!!!\nWrite time to know the Metro time\nWrite help for any help!!!\nWrite exit to quit chatbot\n"
    )

    def __init__(self):
        self.alienbabble = {'Book_tickets': r'.*\s*book.*',
                            'time_tickets': r'time.*',
                            'Help_ticket': r'.*help.*'
                            }

    def greet(self):
        self.name = input("Can I know your name?\n")
        will_help = input(
            f"Hi {self.name}, I'm Kavya. I am Delhi Metro Chatbot. Press 1 to continue!!!\n")
        if will_help in self.negative_responses:
            print("Thanks for visiting chatbot!!!\nHope to see you again!!!")
            return
        self.chat()

    def make_exit(self, reply):
        for command in self.exit_commands:
            if reply == command:
                print("Thanks for visiting chatbot!!!\nHope to see you again!!!")
                return True

    def chat(self):
        reply = input(random.choice(self.random_questions)).lower()
        while not self.make_exit(reply):
            reply = input(self.match_reply(reply))

    def match_reply(self, reply):
        for key, value in self.alienbabble.items():
            intent = key
            regex_pattern = value
            found_match = re.match(regex_pattern, reply)
            if found_match and intent == 'Book_tickets':
                return self.Book_tickets()
            elif found_match and intent == 'time_tickets':
                return self.time_tickets()
            elif found_match and intent == 'Help_ticket':
                return self.Help_ticket()
        if not found_match:
            return self.no_match_intent()

    # The function to search for a station and return its details
    def find_station(self, station_name):
        # Use case-insensitive search to find the station name in the DataFrame
        matches = metro['Station'].str.match(station_name, case=False)
        if matches.any():
            return metro[matches]['Station'].iloc[0]
        return None

    def Book_tickets(self):
        # ONBOARDING STATION
        print("Enter your Onboarding Station:- ")
        search2 = input()
        search1 = search2.lower()
        stat1 = search_line(search1).to_string()

        # DESTINATION STATION
        print("Enter your Destination Station:- ")
        search2 = input()
        search3 = search2.lower()
        stat2 = search_line(search3).to_string()

        # ONBOARDING LINE
        c = metro.loc[search1, "Line"]

        # DESTINATION LINE
        d = metro.loc[search3, "Line"]

        a = metro.loc[search3, "Index"]

        b = metro.loc[search1, "Index"]

        # COUNT
        print("Confirm Ticket!!!")
        print("Write YES for confirmation!!!")
        confirm = input()
        if(confirm == "YES" or confirm == "yes" or confirm == "Yes"):
            if(c == d):
                if(a >= b):
                    count = a - b
                    cost = 0
                    if(count == 0):
                        print("INVALID")
                    elif(count == 1):
                        cost = 10
                    else:
                        cost = 10 + (count-1) * 5
                    print("Your cost for the ticket is:- ", cost)
                else:
                    count = b - a
                    cost = 0
                    if(count == 0):
                        print("INVALID")
                    elif(count == 1):
                        cost = 10
                    else:
                        cost = 10 + (count-1) * 5
                    print("Your cost for the ticket is:- ", cost)

            else:
                print("Different Lines!!!")
                print("Press exit to come out or else select other options to explore\n")
                return
        else:
            print("THANKS FOR VISITING!!!")

        print("How would you like to pay?\n1. UPI\n2. CARDS\n3. INTERNET BANKING\n")
        choice = int(input())
        if(choice == 1):
            print("Your ticket is confirmed!!!\n QR CODE will be generated soon!!!\n")
        elif(choice == 2):
            print("Your ticket is confirmed!!!\n QR CODE will be generated soon!!!\n")
        elif(choice == 3):
            print("Your ticket is confirmed!!!\n QR CODE will be generated soon!!!\n")
        else:
            print("Retry!!!")
        responses = ("Your ticket is confirmed!!!\nPress exit to come out or else select other options to explore\n",
                     "CONFIRMED TICKET!!!\nPress exit to come out or else select other options to explore\n",
                     "Thanks for confirming the ticket!!!\nPress exit to come out or else select other options to explore\n")

        return random.choice(responses)
        count = 0
        cost = 0

    def time_tickets(self):
        # ONBOARDING STATION
        print("Enter your Onboarding Station:- ")
        search2 = input()
        search1 = search2.lower()
        stat1 = search_line(search1).to_string()

        # DESTINATION STATION
        print("Enter your Destination Station:- ")
        search2 = input()
        search3 = search2.lower()
        stat2 = search_line(search3).to_string()

        # ONBOARDING LINE
        c = metro.loc[search1, "Line"]

        # DESTINATION LINE
        d = metro.loc[search3, "Line"]

        a = metro.loc[search3, "Index"]

        b = metro.loc[search1, "Index"]

        if(c == d):
            r2 = random.randint(1, 12)
            r3 = random.randint(0, 59)
            print("Your train will arrive at", r2, ":", r3, "pm")
        else:
            print("Different Lines!!!")

        responses = ("Press exit to come out or select other options to proceed\n",
                     "Press exit to come out or do select other options to proceed\n",
                     "Press exit to come out or choose other options to proceed\n")
        return random.choice(responses)

    def Help_ticket(self):
        print("Q.1 What to do if I paid money and QR Code is not generated?\n",
              "ANS:- Do wait for some time or else write ticket in chatbot !!!\nIf still not resolved contact the nearest metro administrator and do keep the reference number.\n\n",
              "Q.2 How to book tickets?\n",
              "ANS:- Click on Delhi Metro Chatbot.\nThen click any number.\nThen write book.\nThen enter Onboarding station and Destination station and confirm yes.\nYour cost for travel is generated.\nThen enter your mode of payment.\nYour QR Code is generated.\n\n",
              "Q.3 Can you book metro in different lines?\n",
              "ANS:- No. You can book a metro on a single line!!!\n\n",
              "Q.4 How to check time for any metro?\n",
              "ANS:- Click on Delhi Metro Chatbot.\nThen click any number.\nThen write time.\nThen enter Onboarding station and Destination station.\nYour time will be shown\n\n",
              "Q.5 Can you see the time on different lines?\n",
              "ANS:- No. You can check metro timing on a single line!!!\n\n")
        responses = ("For any further queries!!!\nPlease do contact metro help desk!!!\nPress exit to come out or select other options to proceed",
                     "For any further queries!!!\nYou can go to the help center for help!!!\nPress exit to come out or select other options to proceed\n",
                     "For any further queries!!!\nYou can contact Metro help administrator!!!\nPress exit to come out or select other options to proceed\n"
                     )
        return random.choices(responses)

    def no_match_intent(self):
        responses = (
            "No matching options!!!\nPress exit to come out or else select other options to explore\n",
            "Select correct options!!!\nPress exit to come out or else select other options to explore\n", "Do write the name correctly!!!\nPress exit to come out or else select other options to explore\n",
            "Incorrect name!!!\nPress exit to come out or else select other options to explore\n")
        return random.choice(responses)

# Read the data from the CSV file into a pandas DataFrame
metro = pd.read_csv("/content/DELHI_METRO_DATA (2).csv")

# Convert all station names to lowercase for case-insensitive searching
metro['Station'] = metro['Station'].apply(str.lower)

# Set the "Station" column as the DataFrame's index for faster lookups
metro.set_index("Station", inplace=True)

# Function to search for a station and return its details
def search_line(search1):
    result = metro.loc[search1]
    return result

def search_line(search3):
    result = metro.loc[search3]
    return result

# Instantiate the AlienBot class and start the chat
Alien_bot = AlienBot()
Alien_bot.greet()
```
